Importan datos de altura

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def importarAltura4v():
    alt = pd.read_excel('Datos.xlsx', sheet_name=0, header=3, usecols=[3,4], nrows=12)
    return alt


def importarAltura15k():
    alt = pd.read_excel('Datos.xlsx', sheet_name=1, header=3, usecols=[3,4], nrows=8)
    return alt

def importarAltura12k():
    alt = pd.read_excel('Datos.xlsx', sheet_name=2, header=3, usecols=[3,4], nrows=9)
    return alt

def importarAltura7k():
    alt = pd.read_excel('Datos.xlsx', sheet_name=3, header=3, usecols=[3,4], nrows=8)
    return alt

alt4v = importarAltura4v()
alt15k = importarAltura15k()
alt12k = importarAltura12k()
alt7k = importarAltura7k()

```

Graficar alturas

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
fig3, ax3 = plt.subplots()
fig4, ax4 = plt.subplots()

listaEjes = [ax1, ax2, ax3, ax4]

ax1.plot(np.array(alt4v['Tiempo']), np.array(alt4v['Altura']), label = 'Altura de la pesa 15kg, led 4v')
ax2.plot(np.array(alt15k['Tiempo']), np.array(alt15k['Altura']), label = 'Altura de la pesa 15kg, led 2v')
ax3.plot(np.array(alt12k['Tiempo']), np.array(alt12k['Altura']), label = 'Altura de la pesa 12kg, led 2v')
ax4.plot(np.array(alt7k['Tiempo']), np.array(alt7k['Altura']), label = 'Altura de la pesa 7kg, led 2v')

for eje in listaEjes:
    eje.set_xlabel('Tiempo $(s)$')
    eje.set_ylabel('Altura $(cm)$')
    eje.legend()
plt.show()
```

Graficar velocidades

```python
vel1 = np.array(alt4v['Altura']) / np.array(alt4v['Tiempo'])
vel2 = np.array(alt15k['Altura']) / np.array(alt15k['Tiempo'])
vel3 = np.array(alt12k['Altura']) / np.array(alt12k['Tiempo'])
vel4 = np.array(alt7k['Altura']) / np.array(alt7k['Tiempo'])

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
fig3, ax3 = plt.subplots()
fig4, ax4 = plt.subplots()

listaEjes = [ax1, ax2, ax3, ax4]

ax1.plot(np.array(alt4v['Tiempo']), vel1, label = 'Velocidad de la pesa 15kg, led 4v')
ax2.plot(np.array(alt15k['Tiempo']), vel2, label = 'Velocidad de la pesa 15kg, led 2v')
ax3.plot(np.array(alt12k['Tiempo']), vel3, label = 'Velocidad de la pesa 12kg, led 2v')
ax4.plot(np.array(alt7k['Tiempo']), vel4, label = 'Velocidad de la pesa 7kg, led 2v')

for eje in listaEjes:
    eje.set_xlabel('Tiempo $(s)$')
    eje.set_ylabel('Velocidad $(\dfrac{cm}{s})$')
    eje.legend()
plt.show()
```

Importar datos de potencia electrica

Graficar energa

```python
e1 = []
e2 = []
e3 = []
e4 = []

for i in range(len(tiempo4v)):
    e1 += [np.trapz(pot1[0:i], tiempo4v[0:i])]

for i in range(len(tiempo15k)):
    e2 += [np.trapz(pot2[0:i], tiempo15k[0:i])]

for i in range(len(tiempo12k)):
    e3 += [np.trapz(pot3[0:i], tiempo12k[0:i])]

for i in range(len(tiempo7k)):
    e4 += [np.trapz(pot4[0:i], tiempo7k[0:i])]

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
fig3, ax3 = plt.subplots()
fig4, ax4 = plt.subplots()

listaEjes = [ax1, ax2, ax3, ax4]

ax1.plot(tiempo4v, e1, label = 'Energa de la pesa 15kg, led 4v')
ax2.plot(tiempo15k, e2, label = 'Energa de la pesa 15kg, led 2v')
ax3.plot(tiempo12k, e3, label = 'Energa de la pesa 12kg, led 2v')
ax4.plot(tiempo7k, e4, label = 'Energa de la pesa 7kg, led 2v')



for eje in listaEjes:
    eje.set_xlabel('Tiempo $(s)$')
    eje.set_ylabel('Energa $(J)$')
    eje.legend()
plt.show()
```

Graficar cosnervacion de la energia

```python
def encontrarAltura(tiempo, alturas, tiempos):
    alt  = np.interp(tiempo, tiempos, alturas)
    return alt

ep1 = []
ep2 = []
ep3 = []
ep4 = []

def calcularPerdidas(ep, e):
    perdidas = []
    for i in range(len(ep)):
        perdidas += [ep[0]-e[i]-ep[i]]
    return perdidas



def recorridoIterativo(tiempos, alturas, temps, peso):
    ep = []
    for i in tiempos:
        ep += [(alturas[-1] - encontrarAltura(i, alturas, temps)) * 9.8 * peso *0.01]
    return ep

ep1 = recorridoIterativo(tiempo4v, np.array(alt4v['Altura']), np.array(alt4v['Tiempo']), 15)
perdidas1 = calcularPerdidas(ep1, e1)


fig1, ax1 = plt.subplots()
ax1.plot(tiempo4v, ep1, label='Energa Potencial')
ax1.plot(tiempo4v, perdidas1, label='Perdidas Energeticas')
ax1.plot(tiempo4v, e1, label='Energa elctrica generada')
ax1.set_ylabel('Energa $(J)$')
ax1.set_xlabel('Tiempo $(s)$')
ax1.set_title('Transformacin de energa LED 4v, 15kg')
ax1.legend()
plt.show()
```

Graficar resistencia

```python
r1 = voltaje4v/corriente4v
T1 = r1<np.inf
R1 = []
r2 = voltaje15k / corriente15k
T2 = r2<np.inf
R2 = []
r3 = voltaje12k / corriente12k
T3 = r3 < np.inf
R3 = []
r4 = voltaje7k / corriente7k
T4 = r4 < np.inf
R4 = []
volt1 = []
volt2 = []
volt3 = []
volt4 = []
tiempoR4 = []
tiempoR3 = []
tiempoR2 = []
tiempoR1 = []
contador = 0
for val in T1:
    if val:
        R1 += [r1[contador]]
        tiempoR1 += [tiempo4v[contador]]
        volt1 += [voltaje4v[contador]]
    contador += 1
contador = 0
for val in T2:
    if val:
        R2 += [r2[contador]]
        tiempoR2 += [tiempo15k[contador]]
        volt2 += [voltaje15k[contador]]
    contador += 1
contador = 0
for val in T3:
    if val:
        R3 += [r3[contador]]
        tiempoR3 += [tiempo12k[contador]]
        volt3 += [voltaje12k[contador]]
    contador += 1
contador = 0
for val in T4:
    if val:
        R4 += [r4[contador]]
        tiempoR4 += [tiempo7k[contador]]
        volt4 += [voltaje7k[contador]]
    contador += 1
volt4 = np.array(volt4)
volt3 = np.array(volt3)
volt2 = np.array(volt2)
volt1 = np.array(volt1)
R4 = np.array(R4)
R3 = np.array(R3)
R2 = np.array(R2)
R1 = np.array(R1)
tiempoR1 = np.array(tiempoR1)
tiempoR2 = np.array(tiempoR2)
tiempoR3 = np.array(tiempoR3)
tiempoR4 = np.array(tiempoR4)
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
fig3, ax3 = plt.subplots()
fig4, ax4 = plt.subplots()
lisEje = [ax1, ax2, ax3, ax4]
ax1.plot(tiempoR1, R1, label = 'Resistencia para 15kg, led 4v')
ax2.plot(tiempoR2, R2, label = 'Resistencia para 15kg, led 2v')
ax3.plot(tiempoR3, R3, label = 'Resistencia para 12kg, led 2v')
ax4.plot(tiempoR4, R4, label = 'Resistencia para 7kg, led 2v')
for eje in lisEje:
    eje.set_xlabel('Tiempos $(s)$')
    eje.set_ylabel('Resistencia $(\Omega)$')
    eje.legend()
plt.show()
```

```python

def graficar(indexado, volt, r):
    voltGraf = []
    rGraf = []
    for i in indexado:
        voltGraf += [volt[i]]
        rGraf += [r[i]]
    voltGraf = np.array(voltGraf)
    rGraf = np.array(rGraf)
    fig, ax = plt.subplots()
    ax.plot(voltGraf, rGraf, label = 'Resistencia en funcion del voltaje')
    ax.set_ylabel('Resistencia $(\Omega)$')
    ax.set_xlabel('Voltaje $(v)$')
    ax.legend()
    plt.show()

indexado = np.argsort(volt1)
graficar(indexado, volt1, R1)
indexado = np.argsort(volt2)
graficar(indexado, volt2, R2)
indexado = np.argsort(volt3)
graficar(indexado, volt3, R3)
indexado = np.argsort(volt4)
graficar(indexado, volt4, R4)
```
